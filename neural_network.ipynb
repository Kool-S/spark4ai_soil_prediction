```python
#Sensor 1 and all the target variables
y=train_data.iloc[:,1:7]
x_sensor1=train_data.iloc[:,7:23]
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a Min-Max scaler
scaler = MinMaxScaler()

# Fit and transform the data using the scaler
y_scaled_data = scaler.fit_transform(y)
x_sensor1_scaled_data = scaler.fit_transform(x_sensor1)

y_scaled = pd.DataFrame(y_scaled_data, columns=y.columns)
x_sensor1_scaled = pd.DataFrame(x_sensor1_scaled_data, columns=x_sensor1.columns)
```

```python
#separating training and validation dataset (70% for training and 30% for validation)
y_training=y_scaled.iloc[:int(250*0.7)]
y_validation=y_scaled.iloc[int(250*0.7):]
x_sensor1_training=x_sensor1_scaled.iloc[:int(250*0.7)]
x_sensor1_validation=x_sensor1_scaled.iloc[int(250*0.7):]
x_sensor1_validation.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(16,)),  # Input layer
    keras.layers.Dense(48, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.01)),  # Hidden layer with ReLU activation
    keras.layers.Dense(18, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.01)),  # Additional hidden layer(s) if needed
    keras.layers.Dense(6)  # Output layer with six neurons for the six target variables
])

```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x_sensor1_training, y_training, epochs=60, batch_size=20)
```

```python
test_loss = model.evaluate(x_sensor1_validation, y_validation)
print(f'Test Loss (MSE): {test_loss}')
```

```python
def calculate_rmse(actual_values, predicted_values):

    # Calculate squared differences
    squared_differences = (actual_values - predicted_values) ** 2

    # Calculate mean squared difference (MSE)
    mse = np.mean(squared_differences)

    # Calculate RMSE
    rmse = np.sqrt(mse)

    return rmse
```
